```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/gsea_mrna.txt", sep="\t")
df.head()
```

```python
# basic data manipulation
df.dropna(subset=["Hugo_Symbol"], inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop("Entrez_Gene_Id", axis=1, inplace=True)
df.head()
```

```python
# data pivotting
df = df.set_index('Hugo_Symbol').T.reset_index()
df.columns.name = None
df = df.rename(columns={'index': 'Samples'})
df.head()
```

```python
# nan checking
print(f"n rows: {len(df)}")
print(f"n cols with nan: {len(df.isna().sum()[df.isna().sum() > 0])}")
print(f"n cols with 95% or less nan: {len(df.isna().sum()[(df.isna().sum() > 0) & (df.isna().sum() <= len(df) * 0.95)])}")
print("all missing cols have 95% or more nan, so we can drop these cols.")
df.dropna(axis=1, inplace=True)
df.head()
```

```python
# dtype checking
for i, j in df.dtypes.items():
  if j != 'float64':
    print(f"column '{i}' is not float64 type column")
```

```python
samples = df["Samples"]
df.drop("Samples", axis=1, inplace=True)
df.head()
```

```python
from scipy.stats import skew, kurtosis

def check_normal(df):
  results = []

  for col in df.columns:
      s = skew(df[col])
      k = kurtosis(df[col], fisher=False)
      results.append((col, s, k))

  df_norm_check = pd.DataFrame(results, columns=['gene', 'skewness', 'kurtosis'])
  df_norm_check["skewness"] = pd.to_numeric(df_norm_check["skewness"], errors="coerce")
  df_norm_check["kurtosis"] = pd.to_numeric(df_norm_check["kurtosis"], errors="coerce")

  nan_rows = df[df.isna().any(axis=1)].index
  print(nan_rows)

  df_norm_check["non_normal"] = (df_norm_check['skewness'].abs() > 1) | (df_norm_check['kurtosis'] > 5)
  print(f"{round((len(df_norm_check[df_norm_check['non_normal'] == True]) / len(df_norm_check)) * 100, 2)}% of the data is non normal")
```

```python
# normality checking
check_normal(df)
```

```python
# applying Yeo-Johnson Transformation to further normalize data
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson')
df_norm = pd.DataFrame(pt.fit_transform(df), columns=df.columns)

check_normal(df_norm)
```
